```python
import pickle
```

```python
with open('description.pkl', 'rb') as f:
    a = pickle.load(f)
```

```python
def flatten(d):
    this = []
    if d.get('type', None) == 'group':
        for c in d['children']:
            this.extend(flatten(c))
    else:
        this.extend([d])
    return this

def properties(d):
    return [c['name'] for c in flatten(d)]
```

```python
b = properties(a)
b.sort()
b
```
