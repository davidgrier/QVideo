```python
import pickle
```

```python
with open('description.pkl', 'rb') as f:
    a = pickle.load(f)
```

```python
def flatten(d):
    result = []
    if d.get('type', None) == 'group':
        for c in d['children']:
            result.extend(flatten(c))
    else:
        result.extend([d])
    return result
```

```python
b = flatten(a)
```

```python
controls = ['ReverseX', 'ReverseY']
c = [d for d in flatten(a) if d['name'] in controls]
```
